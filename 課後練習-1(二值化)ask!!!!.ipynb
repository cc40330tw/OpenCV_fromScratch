```python
import cv2
import numpy as np

image = cv2.imread("./sample/01-color_space/lena256rgb.jpg")
cv2.imshow("Normal", image)
#cv2.waitKey(0)

# Convert BGR to Gray
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
#cv2.waitKey(0)

#===============RGB二值化(將影像以強度(threshold)區分)=============
#cv2.threshold(gray,thresh,maxval,type)
ret, binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY) #cv2.THRESH_BINARY是轉成黑白；cv2.THRESH_BINARY_INV是反向轉換；cv2.THRESH_TRUNC則是超過刪除
cv2.imshow("Binary", binary)

ret, binary_inv = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV) #cv2.THRESH_BINARY是轉成黑白；cv2.THRESH_BINARY_INV是反向轉換；cv2.THRESH_TRUNC則是超過刪除
cv2.imshow("Binary_INV", binary_inv)

ret, trunc = cv2.threshold(gray,127,255,cv2.THRESH_TRUNC) #cv2.THRESH_BINARY是轉成黑白；cv2.THRESH_BINARY_INV是反向轉換；cv2.THRESH_TRUNC則是超過刪除
cv2.imshow("Trunc", trunc)
#=============================================================

# Convert BGR to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV", hsv)
#cv2.waitKey(0)

# Convert HSV to RGB
bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
cv2.imshow("BGR", bgr)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
image = cv2.imread("octopus.jpg")
hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

#只取出紫色的部分(紫色章魚會變成白色，其餘部分會變成黑色)
lower = np.array([141,0,0])
upper = np.array([164,145,197])
binary = cv2.inRange(hsv,lower,upper)

#(紫色章魚會變成黑色，其餘部分會變成白色)
Bitwise_Not = cv2.bitwise_not(binary) 

#(紫色章魚會變成綠色，其餘部分會變成黑色)
#BitwiseNot = cv2.bitwise_not(image,mask=binary)

#(紫色章魚會變成紫色，其餘部分會變成黑色)原圖與遮罩相疊
#黑AND顏色=黑色；白色AND顏色=顏色
Bitwise_And = cv2.bitwise_and(image,image,mask=binary)

#Bitwise_And2 = cv2.bitwise_and(image,image,mask=Bitwise_Not)

cv2.imshow("Octopus",image)
cv2.imshow("Octopus_hsv",hsv)
cv2.imshow("Octopus_binary",binary)
cv2.imshow("Octopus_Bitwise_Not",Bitwise_Not)

cv2.imshow("BitwiseNot",BitwiseNot)

cv2.imshow("Octopus_Bitwise_And",Bitwise_And)
#cv2.imshow("Octopus_Bitwise_And2",Bitwise_And2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
